## Batch search

```python
from pynoma import GeneSearch
import pandas as pd

searches = [GeneSearch(3,"idua"), GeneSearch(3,"ace2"), GeneSearch(3,"mmp11"), GeneSearch(3,"id4")]
```

```python
datasets=[]
for s in searches:
    s_df, _ = s.get_data() 
    datasets.append(s_df)
```

```python
pd.concat(datasets).fillna(0)
```
